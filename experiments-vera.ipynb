```
import matplotlib.pyplot as plt
```

```
import cv2
import argparse
import numpy as np

def differente(x):
  return (abs(x[0]-x[1])+abs(x[1]-int(x[2]))+abs(int(x[2])-int(x[0])))//3

image = cv2.imread('./img1.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
image.shape
```

```

for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    if differente(image[i][j]) > (15/255):
      image[i][j] = np.array([1, 1, 1])
    if image[i][j] < 0.5:
      image[i][j] = 0
    else:
      image[i][j] = 1
```

```
plt.imshow(image, cmap='gray')
```

```
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = (1 - gray) / 255


for i in range(gray.shape[0]):
  for j in range(gray.shape[1]):
    if gray[i][j] < 0.5:
      gray[i][j] = 0
    else:
      gray[i][j] = 1

plt.imshow(gray, cmap='binary')
plt.axis('off')
```
