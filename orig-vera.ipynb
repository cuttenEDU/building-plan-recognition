```
import matplotlib.pyplot as plt
```

```
import cv2
import argparse
import numpy as np

def differente(x):
  return (abs(int(x[0])-int(x[1]))+abs(int(x[1])-int(x[2]))+abs(int(x[2])-int(x[0])))//3

image = cv2.imread('img1.jpg')
print(image.shape)
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    if differente(image[i][j]) > 15:
      image[i][j] = np.array([255, 255, 255])
```

```
image.shape
```

```

plt.imshow(image, cmap='binary')
plt.axis('off')
```

```
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray = (1 - gray) / 255


for i in range(gray.shape[0]):
  for j in range(gray.shape[1]):
    if gray[i][j] < 0.5:
      gray[i][j] = 0
    else:
      gray[i][j] = 1

plt.imshow(gray, cmap='binary')
plt.axis('off')
```
